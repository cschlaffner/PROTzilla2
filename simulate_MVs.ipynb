### create datasets with simulated MVs

```python
probabilities = [0,0.05,0.1,0.15,0.2,0.25]
```

```python
df.info()
```

```python
for p in probabilities:
    np.random.seed(42)
    na_mask = np.random.choice([True, False], size=df.shape, p=[p,1-p] )
    print(na_mask)
    na_df = df.mask(na_mask,other=np.nan)
    na_df["Intensity"] = [np.nan for i in range(0,482)]
    na_df["iBAQ"] = [np.nan for i in range(0,482)]
    na_df["iBAQ peptides"] = [np.nan for i in range(0,482)]
    na_df["Gene names"] = ["Gene1" for i in range(0,482)]
    na_df.index.name = "Protein IDs"
    na_df.to_csv(f"{PROJECT_PATH}\\user_data\\data\\ba_39_small_{p}.csv",sep="\t")
```
